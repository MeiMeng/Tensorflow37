### 边界填充

```python
t,b,l,r = (50,50,50,50)  #假设上下左右要填充的宽度为50px
replicate = cv2.copyMakeBorder(img, t,b,l,r, borderType=cv2.BORDER_REPLICATE)      #拷贝边界，图片外围1像素拷贝50次
reflect = cv2.copyMakeBorder(img, t,b,l,r, borderType=cv2.BORDER_REFLECT)          #反射边界，图片外围50像素以各边框(0px)为镜面反射
reflect_101 = cv2.copyMakeBorder(img, t,b,l,r, borderType=cv2.BORDER_REFLECT_101)  #反射101边界，图片最外围1px里的49px，以各边最外那1px为镜面反射
wrap = cv2.copyMakeBorder(img, t,b,l,r, borderType=cv2.BORDER_WRAP)                #包裹边界，上50px做下边框，右50px做左边框，対映类推
constant = cv2.copyMakeBorder(img, t,b,l,r, borderType=cv2.BORDER_CONSTANT, value=0)     #常量边界，纯色边框
```

##### 假设上下左右要填充的宽度为50px：
- BORDER_REPLICATE：图片外围1像素拷贝50次：aaaaaa abcdef ffffff
- BORDER_REFLECT：图片外围50像素以各边框(0px)为镜面反射：fedcba abcdef fedcba
- BORDER_REFLECT_101：图片最外围1px里的49px，以各边最外那1px为镜面反射：fedcb abcdef edcba
- BORDER_WRAP：上50px做下边框，右50px做左边框，対映类推：def abcdef abc
- BORDER_CONSTANT：纯色边框。

```python
#图片按RGB重组
def reSort2RGB(img):
    b,g,r=cv2.split(img)
    return cv2.merge([r,g,b])

img1 = reSort2RGB(img)
replicate1 = reSort2RGB(replicate)
reflect1 = reSort2RGB(reflect)
reflect_1011 = reSort2RGB(reflect_101)
wrap1 = reSort2RGB(wrap)
constant1 = reSort2RGB(constant)

plt.figure(figsize=(12, 8)) 
plt.subplot(231), plt.imshow(img1),          plt.title('original')
plt.subplot(232), plt.imshow(replicate1),    plt.title('replicate')
plt.subplot(233), plt.imshow(reflect1),      plt.title('reflect')
plt.subplot(234), plt.imshow(reflect_1011),  plt.title('reflect_101')
plt.subplot(235), plt.imshow(wrap1),         plt.title('wrap')
plt.subplot(236), plt.imshow(constant1),     plt.title('constant')
```


### 图像的叠加与融合
无论何种形式的叠加，前提都是：两个矩阵shape相同。


#### 直接相加

```python
(gao+lena)[:5, :, 0]    #超出255的像素被自动%255
```

```python
cv_show('jia', (gao+lena))    #这个图像都没法看了
```

#### 使用OpenCV相加

```python
cv2.add(gao,lena)[:5, :, 0]   #超出255的像素都记为255
```

```python
cv_show('add', cv2.add(gao,lena))   #亮度叠加，这个图像更合理
```

#### 图像融合

```python
res = cv2.addWeighted(gao, 0.6,  lena, 0.4,  0)    #按权重相加，最后提亮一个常数值： R=αX1+βX2+b  (注：α+β=1)
```

```python
res[:5, :, 0]
```

```python
cv_show('res', res)         #图像按权重融合，融合效果，不错！
```
