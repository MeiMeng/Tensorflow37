```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

from IPython.display import Latex

from IPython.display import HTML
```

```python
def cv_show(name, img):
    cv2.imshow(name, img)
    cv2.waitKey(0)  #0表示任意键，其他数字表示等待ms数
    cv2.destroyAllWindows()
```

```python
#图片按RGB重组
def reSort2RGB(img):
    b,g,r=cv2.split(img)
    return cv2.merge([r,g,b])


#绘图pic
def plt_showRGB( img,  size, title, s=True ):
    if (len(img) > 1) and (len(img) < 20):
        img = np.hstack( img )  #横向拼接
    img = reSort2RGB(img)       #按RGB重组
    if s:
        plt.figure(figsize=size) 
        plt.axis('off')
    plt.imshow(img),  plt.title(title)
    
def plt_showGray( img,  size, title ):
    if (len(img) > 1) and (len(img) < 20):
        img = np.hstack( img )  #横向拼接
    plt.figure(figsize=size) 
    plt.axis('off')
    plt.imshow(img, 'gray'),  plt.title(title)
```

#### dst	=	cv.cvtColor(	src, code  [, dst[, dstCn]]	)
- cv.COLOR_BGR2RGB
- cv.COLOR_BGR2GRAY

```python
gao = cv2.imread('gao.jpg')
gray = cv2.cvtColor(gao, cv2.COLOR_BGR2GRAY)

plt_showGray( gray, (10,8), 'gray' )
```


# 图像阈值
### 二值化


####  retval, dst	=	cv2.threshold(	src, thresh, maxval, type [, dst]	)
#### Parameters
- **src**:	&nbsp;&nbsp; 输入数组（只能一个多通道）。 input array.	
- **dst**:	&nbsp;&nbsp; 与src具有相同大小，类型和相同通道数的输出数组。 output array of the same size and type and the same number of channels as src. 
- **thresh**:	&nbsp;&nbsp; 阈值。 threshold value. 
- **maxval**:	&nbsp;&nbsp; THRESH_BINARY和THRESH_BINARY_INV阈值类型使用的最大值。 
    maximum value to use with the THRESH_BINARY and THRESH_BINARY_INV thresholding types.   
- **type**:	    &nbsp;&nbsp; 二值化操作的类型。 thresholding type. 
    - cv2.THRESH_BINARY 
    - cv2.THRESH_BINARY_INV 
    - cv2.THRESH_TRUNC 
    - cv2.THRESH_TOZERO 
    - cv2.THRESH_TOZERO_INV 
    - cv2.THRESH_MASK 
    - cv2.THRESH_OTSU 
    - cv2.THRESH_TRIANGLE 

![jupyter](./thresholdType3.png)

```python
retval1, bina = cv2.threshold( gray, 127, 255, cv2.THRESH_BINARY )
retval2, bina_INV = cv2.threshold( gray, 127, 255, cv2.THRESH_BINARY_INV )
retval3, trunc = cv2.threshold( gray, 127, 255, cv2.THRESH_TRUNC )
retval4, tozero = cv2.threshold( gray, 127, 255, cv2.THRESH_TOZERO )
retval5, tozero_INV = cv2.threshold( gray, 127, 255, cv2.THRESH_TOZERO_INV )
retval6, mask = cv2.threshold( gray, 127, 255, cv2.THRESH_MASK )
retval7, otsu = cv2.threshold( gray, 127, 255, cv2.THRESH_OTSU )
retval8, triangle = cv2.threshold( gray, 127, 255, cv2.THRESH_TRIANGLE )
```

```python
plt.figure(figsize=(20, 7))   #整体画图的 宽高比 (尺寸*100px)
plt.subplot(241), plt.imshow(bina,     'gray'),    plt.title('bina')
plt.subplot(242), plt.imshow(bina_INV, 'gray'),    plt.title('bina_INV')
plt.subplot(243), plt.imshow(tozero,   'gray'),    plt.title('tozero')
plt.subplot(244), plt.imshow(tozero_INV, 'gray'),  plt.title('tozero_INV')
plt.subplot(245), plt.imshow(trunc,    'gray'),    plt.title('trunc')
plt.subplot(246), plt.imshow(mask,     'gray'),    plt.title('mask')
plt.subplot(247), plt.imshow(otsu,     'gray'),    plt.title('otsu')
plt.subplot(248), plt.imshow(triangle, 'gray'),    plt.title('triangle')
```

## 均值滤波

**归一化框滤镜模糊**： 平均卷积操作， 过滤器 3\*3 矩阵中的值都为1，9格数相加后再➗9，得到一个新的矩阵元素。

**dst	=	cv.blur(	src, ksize  [, dst[, anchor[, borderType]]]	)**

- **SRC**	&nbsp;&nbsp;&nbsp;&nbsp;输入图像；它可以具有任意数量的通道，这些通道可以独立处理，但深度应为CV_8U，CV_16U，CV_16S，CV_32F或CV_64F。
- **DST**	&nbsp;&nbsp;&nbsp;&nbsp;输出与src大小和类型相同的图像。
- **ksize**	  &nbsp;&nbsp;&nbsp;&nbsp;内核大小模糊。
- **锚**	&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;锚点 默认值Point（-1，-1）表示锚点位于内核中心。
- **borderType**	&nbsp;&nbsp;&nbsp;&nbsp;用于推断图像外部像素的边框模式，请参见BorderTypes


```python
lena = cv2.imread('./lena1.jpg')
blur = cv2.blur( lena, (3,3) )   #均值滤波
```

```python
#绘图
lena2 = reSort2RGB(lena)    #按 RGB 重组
blur1 = reSort2RGB(blur)   #按 RGB 重组

plt.figure(figsize=(12, 6))             #整体画图的 宽高比 (尺寸*100px)
plt.subplot(121),  plt.imshow(lena2),   plt.title('lena')
plt.subplot(122),  plt.imshow(blur1),   plt.title('blur')
```

## 方框滤波 

#### dst	= cv.boxFilter(	src, ddepth, ksize  [, dst[, anchor[, normalize[, borderType]]]]  )


- **SRC**	&nbsp;&nbsp;&nbsp;&nbsp;输入图像。
- **DST**	&nbsp;&nbsp;&nbsp;&nbsp;输出与src大小和类型相同的图像。
- **ddepth**	&nbsp;&nbsp;&nbsp;&nbsp;输出图像深度（使用src.depth（）时为 -1 ）。
- **ksize**	 &nbsp;&nbsp;&nbsp;&nbsp;内核大小模糊。
- **锚**	    &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;锚点 默认值Point（-1，-1）表示锚点位于内核中心。
- **正常化**	&nbsp;&nbsp;&nbsp;&nbsp;标志，指定是否通过其区域对内核进行规范化。
- **borderType**	&nbsp;&nbsp;&nbsp;&nbsp;用于推断图像外部像素的边框模式，请参见BorderTypes

```python
img = cv2.imread('./lena1.jpg')
box = cv2.boxFilter( img, -1, (3,3), normalize=False )  #默认 normalize=True 等同于均值滤波

#加上 normalize=False ，相当于不➗9, 会发生亮度越界，都变为255


#绘图
img1 = reSort2RGB(img)    #按 RGB 重组
box1 = reSort2RGB(box)   #按 RGB 重组

plt.figure(figsize=(12, 6))             #整体画图的 宽高比 (尺寸*100px)
plt.subplot(121),  plt.imshow(img1),   plt.title('ori')
plt.subplot(122),  plt.imshow(box1),   plt.title('box')
```

```python
normalize = cv2.boxFilter( img, -1, (3,3) )    # 默认 normalize=True 等同于均值滤波


#绘图
img1 = reSort2RGB(img)    #按 RGB 重组
normalize1 = reSort2RGB(normalize)   #按 RGB 重组

plt.figure(figsize=(12, 6))             #整体画图的 宽高比 (尺寸*100px)
plt.subplot(121),  plt.imshow(img1),   plt.title('ori')
plt.subplot(122),  plt.imshow(normalize1),   plt.title('normalize')
```

## 高斯滤波
卷积核数值满足高斯分布，最重视每个单元中最中间的那个数值，核的最中间是1，其他地方小于1。

**DST	=	cv.GaussianBlur（	src，ksize，sigmaX [，dst [，sigmaY [，borderType]]]	）**

- **SRC**   &nbsp;&nbsp;&nbsp;&nbsp;	输入图像；图像可以具有任意数量的通道，这些通道可以独立处理，但深度应为CV_8U，CV_16U，CV_16S，CV_32F或CV_64F。
- **DST**	 &nbsp;&nbsp;&nbsp;&nbsp;输出与src大小和类型相同的图像。
- **ksize**	 &nbsp;&nbsp;&nbsp;&nbsp;高斯核大小。ksize.width和ksize.height可以不同，但​​它们都必须为正数和奇数。或者，它们可以为零，然后根据sigma计算得出。
- **sigmaX**	&nbsp;&nbsp;&nbsp;&nbsp; X方向上的 高斯核 标准偏差。
- **sigmaY**	&nbsp;&nbsp;&nbsp;&nbsp; Y方向的高斯核标准差；如果sigmaY为零，则将其设置为等于sigmaX；如果两个sigmas为零，则分别从ksize.width和ksize.height计算得出（有关详细信息，请参见getGaussianKernel）；为了完全控制结果，而不管将来可能对所有这些语义进行的修改，建议指定所有ksize，sigmaX和sigmaY。
- **borderType**  &nbsp;&nbsp;&nbsp;&nbsp; 像素外推方法，请参见BorderTypes

```python
gauss = cv2.GaussianBlur(img, (5,5), 1)


#绘图
gauss1 = reSort2RGB(gauss)   #按 RGB 重组

plt.figure(figsize=(12, 6))           #整体画图的 宽高比 (尺寸*100px)
plt.subplot(121),  plt.imshow(img1),   plt.title('ori')
plt.subplot(122),  plt.imshow(gauss1),   plt.title('gauss')
```

## 中值滤波
- 使用中值滤镜模糊图像。
- 该函数使用具有𝚔𝚜𝚒𝚣𝚎 × 𝚔𝚜𝚒𝚣𝚎光圈的中值滤镜对图像进行平滑处理。多通道图像的每个通道都是独立处理的。支持就地操作。
- 将3*3个数字从小到大排序，取第5个数字。


**DST	=	cv.medianBlur（	src，ksize [，dst]	）**
- **ksize** &nbsp;&nbsp;&nbsp;&nbsp;	孔径线性尺寸 它必须是奇数且大于1，例如：3、5、7 ...

```python
median = cv2.medianBlur(img, 5)


#绘图
median1 = reSort2RGB(median)   #按 RGB 重组

plt.figure(figsize=(12, 6))             #整体画图的 宽高比 (尺寸*100px)
plt.subplot(121),  plt.imshow(img1),   plt.title('ori')
plt.subplot(122),  plt.imshow(median1),   plt.title('median')
```


## 图像数组拼接


```python
res = np.hstack( (blur, gauss, median) )   # hstack横向拼接，vstack纵向拼接

#绘图
res1 = reSort2RGB(res)
plt.figure(figsize=(18, 6)) 
plt.imshow(res1),  plt.title('blur, gauss, median')
```
