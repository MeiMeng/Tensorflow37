```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

from IPython.display import Latex

from IPython.display import HTML 
```

```python
def cv_show(name, img):
    cv2.imshow(name, img)
    cv2.waitKey(0)  #0表示任意键，其他数字表示等待ms数
    cv2.destroyAllWindows()
    

#图片按RGB重组
def reSort2RGB(img):
    b,g,r=cv2.split(img)
    return cv2.merge([r,g,b])


#图像转置
def transposePic(img):
    img1 = np.zeros( (img.shape[1], img.shape[0], img.shape[2] ) )
    for i in range(0, 2):
        for x in range(0, img.shape[0]):
            for y in range(0, img.shape[1]):
                img1[y][x][i] = img[x][y][i]
    return img1 


#绘图pic
def drawRGB( images,  size, title ):
    res = np.hstack( images )
    res = reSort2RGB(res)
    plt.figure(figsize=size)
    plt.axis('off')
    plt.imshow(res),  plt.title(title)
    
def drawGray( images,  size, title ):
    res = np.hstack( images )
    plt.figure(figsize=size) 
    plt.axis('off')
    plt.imshow(res, 'gray'),  plt.title(title)
```

#### edges	=	cv.Canny(	image, threshold1, threshold2   [, edges[, apertureSize[, L2gradient]]]	)

- threshold1和threshold2之间的最小值用于边缘链接。最大值用于查找强边的初始段。

#### 参数
- **图片**   &nbsp;&nbsp;&nbsp;&nbsp;	8位输入图像。
- **边缘**	&nbsp;&nbsp;&nbsp;&nbsp;  输出边缘图；单通道8位图像，其大小与image相同。
- **阈值1**  &nbsp;&nbsp;&nbsp;&nbsp;	迟滞过程的第一阈值。
- **阈值2**  &nbsp;&nbsp;&nbsp;&nbsp;	迟滞过程的第二个阈值。
- **apertureSize**  &nbsp;&nbsp;&nbsp;&nbsp; Sobel操作员的光圈尺寸。
- **L2gradient**  &nbsp;&nbsp;&nbsp;&nbsp;	一个标志，指示是否应使用更准确的[ 数学处理错误]规范[ 数学处理错误]来计算图像梯度幅度（L2gradient = true），或者默认[ 数学处理错误]规范[ 数学处理错误]是否为足够（L2gradient = false）。

```python
img = cv2.imread('yuan.jpg')
v1 = cv2.Canny(img, 80, 150)
v2 = cv2.Canny(img, 50, 100)
drawGray( (v1,v2), (30, 8), 'v1, v2' )
```
